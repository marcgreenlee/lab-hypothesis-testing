In this challenge, we will be working with pokemon data. The data can be found here:

```python
#libraries
import pandas as pd
import scipy.stats as st
import numpy as np
import scipy.stats as st
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/pokemon.csv")
df
```

- We posit that Pokemons of type Dragon have, on average, more HP stats than Grass. Choose the propper test and, with 5% significance, comment your findings.

```python
#H0: mean Dhp > mean Ghp 
#H1: mean Dhp <= mean Ghp
alpha = 0.05
import seaborn as sns
grass_df = df[df['Type 1'] == 'Grass']['HP']
dragon_df = df[df['Type 1'] == 'Dragon']['HP']
sns.histplot(data = dragon_df)

dragon_df.count(), grass_df.count()
```

```python
st.ttest_ind(dragon_df, grass_df, equal_var = False, alternative = 'less')
```

The P-value obtained, higher than the alpha value, indicates the data strongly supports the null hypothesis, Dragon Pokemons have a higher HP than Grass Pokemons on average.

- We posit that Legendary Pokemons have different stats (HP, Attack, Defense, Sp.Atk, Sp.Def, Speed) when comparing with Non-Legendary. Choose the propper test and, with 5% significance, comment your findings.


```python
df_legendary = df[df['Legendary'] == True] [['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation']].mean()
df_nonlegendary = df[df['Legendary'] == False] [['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation']].mean()
```

```python
#Set the hypothesis

#H0: mu_legendary = mu_non_legendary
#H1: mu_legendary != mu_non_legendary

#significance level = 0.05

st.ttest_ind(df_legendary,df_nonlegendary, equal_var=False, alternative="two-sided")
```

the P-value being higher than the significance level, implies we do'nt have evidende to reject the null hypothesis, which states taht legendary pokemons, have the same stats than non legendary pokemons, on average. 

**Challenge 2**

In this challenge, we will be working with california-housing data. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv

```python
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/california_housing.csv")
df
```

**We posit that houses close to either a school or a hospital are more expensive.**

- School coordinates (-118, 37)
- Hospital coordinates (-122, 34)

We consider a house (neighborhood) to be close to a school or hospital if the distance is lower than 0.50.

Hint:
- Write a function to calculate euclidean distance from each house (neighborhood) to the school and to the hospital.
- Divide your dataset into houses close and far from either a hospital or school.
- Choose the propper test and, with 5% significance, comment your findings.
 

```python
from scipy.stats import ttest_ind, levene


school_coords = (-118, 37)
hospital_coords = (-122, 34)

# Function to calculate the Euclidean distance
def calculate_distance(house_longitude, house_latitude, reference_coords):
    return np.sqrt((house_longitude - reference_coords[0])**2 + (house_latitude - reference_coords[1])**2)

# Calculate distances from each house to school and hospital
df['distance_to_school'] = df.apply(lambda row: calculate_distance(row['longitude'], row['latitude'], school_coords), axis=1)
df['distance_to_hospital'] = df.apply(lambda row: calculate_distance(row['longitude'], row['latitude'], hospital_coords), axis=1)

# Define 'close' houses: distance < 0.50 to either school or hospital
df['close_to_school_or_hospital'] = (df['distance_to_school'] < 0.50) | (df['distance_to_hospital'] < 0.50)

# Divide the data into two groups: close and far houses
close_houses = df[df['close_to_school_or_hospital'] == True]
far_houses = df[df['close_to_school_or_hospital'] == False]

stat_levene, p_levene = levene(close_houses['price'], far_houses['price'])
print(f"Levene's test p-value: {p_levene}")

# Perform t-test (Welch’s t-test if variances are unequal)
if p_levene < 0.05:
    t_stat, p_value = ttest_ind(close_houses['price'], far_houses['price'], equal_var=False)
else:
    t_stat, p_value = ttest_ind(close_houses['price'], far_houses['price'], equal_var=True)

print(f"t-statistic: {t_stat}, p-value: {p_value}")

# Hypothesis testing decision
if p_value < 0.05:
    print("Reject the null hypothesis: Houses close to a school or hospital are significantly more expensive.")
else:
    print("Fail to reject the null hypothesis: No significant difference in house prices between close and far houses.")

```
